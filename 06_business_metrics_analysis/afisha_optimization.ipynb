## Рассчитаем метрики и построим графики

### Рассчитаем продуктовые метрики

#### Рассчитаем DAU, WAU и MAU и найдём средние значения этих метрик

```python
# добавим столбцы с датой, неделей и месяцем сессии
visits['session_date'] = visits['start_ts'].dt.date
visits['session_week'] = visits['start_ts'].astype(
    'datetime64[W]'
)
visits['session_month'] = visits['start_ts'].astype(
    'datetime64[M]'
)

# рассчитаем метрики
dau_total = visits.groupby('session_date').agg({'uid':'nunique'})
wau_total = visits.groupby('session_week').agg({'uid':'nunique'})
mau_total = visits.groupby('session_month').agg({'uid':'nunique'})
```

Вычислим средние значения этих метрик за весь период

```python
# найдём их средние значения
dau_mean = dau_total.mean()
wau_mean = wau_total.mean()
mau_mean = mau_total.mean()

# выведем значения
print(f'Среднее значение метрики DAU: {int(dau_mean)}')
print(f'Среднее значение метрики WAU: {int(wau_mean)}')
print(f'Среднее значение метрики MAU: {int(mau_mean)}')
```

##### Sticky Factor

Sticky Factor показывает привлекательность проекта с точки зрения монетизации - чем выше вовлечённость пользователей, тем выше лояльность пользователей к нему. Рассчитаем средние значения этих метрик для проекта

```python
# подготовим данные для расчёта
sticky = pd.DataFrame(dau_total).reset_index()
sticky['session_date'] = pd.to_datetime(sticky['session_date'])
sticky['session_week'] = sticky['session_date'].astype(
    'datetime64[W]'
)
sticky['session_month'] = sticky['session_date'].astype(
    'datetime64[M]'
)

wau_total = wau_total['uid']
wau_total.name = 'wau'
mau_total = mau_total['uid']
mau_total.name = 'mau'

sticky = sticky.join(wau_total, on='session_week')
sticky = sticky.join(mau_total, on='session_month')
sticky = sticky.rename(columns = {'uid':'dau'})
```

```python
# рассчитаем метрики
sticky['st_weekly'] = sticky['dau'] / sticky['wau']
sticky['st_monthly'] = sticky['dau'] / sticky['mau']
```

```python
# сохраним данные о значения недельного sticky factor для каждой недели
sticky_weekly = sticky.groupby('session_week').agg({'st_weekly':'mean'}).reset_index()

# оценим значения метрики
sticky_weekly['st_weekly'].describe().T
```

Как мы видим, у нас есть аномальное значение максимума, которое сильно влияет на среднее значение. На него нужно посмотреть более пристально

```python
# выведем строку с этим значением
display(sticky_weekly.sort_values(by='st_weekly', ascending=False).head(1))
```

Когда неделя аномально короткая (в нашем случае состоит из одного дня), то и метрики имеют аномальные значения. Можем удалить эту строчку без сожаления и посмотреть на реальный максимум

```python
# удалим аномалию
sticky_weekly.drop(52, inplace=True)

# выведем реальный максимум
display(sticky_weekly.sort_values(by='st_weekly', ascending=False).head(1))
```

Наибольшее значение недельный sticky factor имел на неделе, которая начиналась 29 марта 2017 года

```python
# выведем минимум
display(sticky_weekly.sort_values(by='st_weekly').head(1))
```

Минимальное значение метрика имела на неделе, которая начиналась 8 октября 2017 года

```python
# рассчитаем среднее значение за весь период
sticky_weekly_mean = sticky_weekly['st_weekly'].mean()

# выведем среднее значение
print('Средний недельный sticky factor составляет {:.2%}'.format(sticky_weekly_mean))
```

Проделаем всё то же самое для месячного sticky factor

```python
# сохраним данные о значения месячного sticky factor для каждого месяца
sticky_monthly = sticky.groupby('session_month').agg({'st_monthly':'mean'}).reset_index()

# оценим значения метрики
display(sticky_monthly['st_monthly'].describe().T)
```

Аномалий как в случае с недельной метрикой не обнаружено. Выведем строку с максимальным значением метрики

```python
# выведем максимум
display(sticky_monthly.sort_values(by='st_monthly', ascending=False).head(1))
```

Наибольшая вовлечённость пользователей была в феврале 2018 года. Посмотрим на минимальное значение

```python
# выведем минимум
display(sticky_monthly.sort_values(by='st_monthly').head(1))
```

Наименьшая вовлечённость - в августе 2017 года.

```python
# рассчитаем среднее значение за весь период
sticky_monthly_mean = sticky_monthly['st_monthly'].mean()

# выведем среднее значение
print('Средний месячный sticky factor составляет {:.2%}'.format(sticky_monthly_mean))
```

##### Рассчитаем метрики в разрезе устройств

```python
# рассчитаем метрики
dau_by_device = visits.groupby(['device', 'session_date']).agg({'uid':'nunique'}).reset_index()
wau_by_device = visits.groupby(['device', 'session_week']).agg({'uid':'nunique'}).reset_index()
mau_by_device = visits.groupby(['device', 'session_month']).agg({'uid':'nunique'}).reset_index()

# переименуем столбцы для удобства
dau_by_device = dau_by_device.rename(columns={'uid':'dau'})
wau_by_device = wau_by_device.rename(columns={'uid':'wau'})
mau_by_device = mau_by_device.rename(columns={'uid':'mau'})

# начнём создавать сводную таблицу
metrics_by_device = dau_by_device

# выделим неделю и месяц сессии
metrics_by_device['session_week'] = metrics_by_device['session_date'].astype(
    'datetime64[W]'
)
metrics_by_device['session_month'] = metrics_by_device['session_date'].astype(
    'datetime64[M]'
)

# добавим недостающие данные в сводную таблицу
metrics_by_device = metrics_by_device.merge(wau_by_device, on=['device', 'session_week'])
metrics_by_device = metrics_by_device.merge(mau_by_device, on=['device', 'session_month'])

# рассчитаем sticky factors в разрезе устройств
metrics_by_device['st_weekly'] = metrics_by_device['dau'] / metrics_by_device['wau']
metrics_by_device['st_monthly'] = metrics_by_device['dau'] / metrics_by_device['mau']
```

```python
# посчитаем средние значение для всего периода по разным устройствам
dau_by_device_mean = dau_by_device.groupby('device').agg({'dau':'mean'}).round().reset_index()
wau_by_device_mean = wau_by_device.groupby('device').agg({'wau':'mean'}).round().reset_index()
mau_by_device_mean = mau_by_device.groupby('device').agg({'mau':'mean'}).round().reset_index()

# посчитаем долю метрик для каждого устройства
dau_by_device_mean['share'] = round(dau_by_device_mean['dau'] / dau_by_device_mean['dau'].sum() * 100, 2)
wau_by_device_mean['share'] = round(wau_by_device_mean['wau'] / wau_by_device_mean['wau'].sum() * 100, 2)
mau_by_device_mean['share'] = round(mau_by_device_mean['mau'] / mau_by_device_mean['mau'].sum() * 100, 2)
```

#### Рассчитаем Retention Rate

Сначала подготовим необходимые данные для когортного анализа

```python
# выделим время первого визита
first_visits = visits.groupby('uid')['start_ts'].min()
first_visits.name = 'first_visit_ts'

# присоединим время первого визита к датасету `visits`
visits = visits.join(first_visits, on='uid')

# выделим месяц из времени первого визита
visits['first_visit_month'] = visits['first_visit_ts'].astype(
    'datetime64[M]'
)

# выделим месяц для времени начала каждой сессии
visits['session_month'] = visits['start_ts'].astype(
    'datetime64[M]'
)

# добавим столбец со временем жизни когорты
visits['lifetime'] = ((
    visits['session_month'] - visits['first_visit_month']
) / np.timedelta64(1, 'M')).round().astype('int')
```

```python
# создаём датасет с когортами
cohorts = visits.groupby(['first_visit_month', 'lifetime']).agg({'uid':'nunique'}).reset_index()

# считаем изначальное количество пользователей в когортах
initial_users_count = cohorts[cohorts['lifetime'] == 0][
    ['first_visit_month', 'uid']
]

# переименовываем столбцы для удобства объединения
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})
```

```python
# добавим столбец с изначальным количеством пользователей к датасету с когортами
cohorts = cohorts.merge(initial_users_count,on='first_visit_month')

# считаем RR
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']
```

```python
# создаём сводную таблицу
retention_pivot = cohorts.pivot_table(
    index='first_visit_month',
    columns='lifetime',
    values='retention',
    aggfunc='sum',
)

# делаем красивые индексы
retention_pivot.index = retention_pivot.index.strftime('%m-%Y')
```

```python
# выведем тепловую карту на основании сводной таблицы
sns.set(style='white', palette="tab10")
plt.figure(figsize=(20, 12))
plt.title('Retention Rate по когортам')
ax=sns.heatmap(
    retention_pivot,
    annot=True,
    fmt='.1%',
    vmax=0.1,
    cmap='viridis_r'
)
ax.set_xlabel('Время жизни когорты')
ax.set_ylabel('Когорта')
plt.show()
```

```python
# рассчитаем среднее значение на 2 месяц жизни когорты
avg_rr = retention_pivot[1].mean()

# выведем на экран
print('Средний Retention Rate на второй месяц жизни когорт равен {:.1%}'.format(avg_rr))
```

```python
# выделим время первой покупки 
first_orders = orders.groupby('uid')['buy_ts'].min()

# переименуем Series
first_orders.name = 'first_buy_ts'

# создаём датасет со сведениями о времени первого визита и первой покупки
orders_visits = pd.DataFrame(first_orders).reset_index()
orders_visits = orders_visits.join(first_visits, on='uid')

# посчитаем, сколько секунд прошло с первого визита до первой покупки 
orders_visits['time_to_buy'] = (orders_visits['first_buy_ts'] - orders_visits['first_visit_ts']).dt.seconds

# посмотрим на распределение
ax = orders_visits['time_to_buy'].hist(
    bins=100,
    figsize=(14,9)
)
ax.set_title(
    'Гистограмма распределения времени с момента первого посещения сайта Яндекс.Афиша до первой покупки (данные за период 06.2017 - 05.2018)'
)
ax.set_xlabel(
    'Время с момента первого посещения до первой покупки, с'
)
plt.show()
```

Мы видим, что распределение отличается от нормального, поэтому такие меры среднего как среднее арифметическое или медиана не применимы для описания среднего времени от первого визита до первой покупки. В данном случае логичным будет применить моду.

```python
# рассчитаем моду
ttb_avg = orders_visits['time_to_buy'].mode()

# выведем на экран
print(f'Средний временной интервал между первым посещением сайта и первой покупкой составляет {int(ttb_avg)} с')
```

#### Рассчитаем среднее количество покупок на одного покупателя за 6 месяцев в 2017 году

```python
# выделим месяц из времени покупки
orders['buy_month'] = orders['buy_ts'].astype(
    'datetime64[M]'
)

# возьмём срез данных за 6 месяцев в 2017 году и сгруппируем по месяцу покупки
# посчитаем для каждого месяца число покупок и количество уникальных покупателей
orders_grouped = orders.query('buy_month < "2018-01-01"').groupby('buy_month').agg({'buy_ts':'count', 'uid':'nunique'}).reset_index()

# рассчитаем количество покупок на пользователя
orders_grouped['purchases_per_user'] = orders_grouped['buy_ts'] / orders_grouped['uid']

# переименуем столбцы
orders_grouped = orders_grouped.rename(columns = {'buy_ts':'total_purchases', 'uid':'users_count'})

# построим график
ax = orders_grouped.plot(
    x='buy_month',
    y='purchases_per_user',
    grid=True,
    legend=False,
    figsize=(14, 9),
    style='-x'
)
ax.set_title(
    'Изменение среднего количества покупок на одного покупателя для сайта Яндекс.Афиша за период с 06.2017 по 12.2017'
)
ax.set_xlabel(
    'Месяц'
)
ax.set_ylabel(
    'Среднее количество покупок на одного покупателя, шт'
)
plt.show()
```

#### Выясним, как меняется LTV на покупателя по когортам

Рассчитаем средний LTV по когортам за 6 месяцев; в расчёт включим когорты, «прожившие» не менее 6 месяцев. Маржинальность сервиса — 100%. Отразим изменения метрики во времени на графике

```python
# выделим время первой покупки 
first_orders = orders.groupby('uid').agg({'buy_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']

# посчитаем количество пользователей в каждой когорте
cohort_sizes = (
    first_orders.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']

# прикрепим размеры когорт к датасету с заказами
orders_new = pd.merge(orders,first_orders, on='uid')

# сформируем датасет с когортами
cohorts = (
    orders_new.groupby(['first_order_month', 'buy_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)

#сформируем итоговый датасет
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
```

```python
# маржинальность 100%
margin_rate = 1

# рассчитаем валовую выручку и возраст когорт
report['gp'] = report['revenue'] * margin_rate
report['age'] = (
    report['buy_month'] - report['first_order_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
```

```python
# рассчитаем ltv
report['ltv'] = report['gp'] / report['n_buyers']

# в итоговый вывод включаем только когорты, просуществовавшие не менее 6 месяцев
# также включаем данные только за первые 6 месяцев существования когорт
output = report.query('(age <= 6) and (first_order_month <= "2017-11-01")').pivot_table(
    index='first_order_month', columns='age', values='ltv', aggfunc='mean'
).cumsum(axis=1).round(2)
```

```python
# проверим, как сформировался итоговый вывод
display(output)
```

```python
# рассчитаем средний ltv по всем когортам за каждый месяц существования; данные сохраним в список
avg_ltv = []

for i in range(0, 7):
     avg_ltv.append(output[i].mean())

# создадим новый датафрейм на основе нашего списка
avg_ltv = pd.DataFrame(avg_ltv, index=list(range(0, 7)))
avg_ltv = avg_ltv.rename(columns={0:'avg_ltv'})


# построим график
ax = avg_ltv.plot(
    kind='bar',
    figsize=(20, 12),
    y='avg_ltv',
    legend=False,
    grid=True
)
ax.set_title(
    'Изменение среднего LTV по времени жизни когорты за 6 месяцев (данные за 06.2017-05.2018 для когорт с временем жизни не менее 6 мес.)'
)
ax.set_xlabel(
    'Время жизни когорты, мес.'
)
ax.set_ylabel(
    'Средний LTV, у.е.'
)
plt.show()
```

Теперь нужно посчитать средний CAC для каждого источника трафика. Для этого разделим общие затраты на данный источник на число уникальных пользователей, привлечённых источником

```python
# так как в сведениях о покупках не отражён источник трафика, выделим первый источник трафика для каждого пользователя
first_sources = visits.groupby('uid').agg({'source_id':'first'})

# присоединим эти сведения к данным о покупках
orders_with_sources = orders.join(first_sources, on='uid')
```

Исследуем изменение CAC по времени в разрезе источников трафика

```python
# сгруппируем расходы по источнику трафика
costs_by_source = costs.groupby('source_id').agg({'costs':'sum'}).reset_index()

#
orders_with_sources = orders_with_sources.drop(columns=['buy_ts', 'revenue', 'buy_date'], axis=1)

# сгруппируем расходы по источнику трафика и месяцу и сохраним их в отдельном датасете
costs_grouped_new = costs.groupby(['source_id', 'cost_month']).agg({'costs':'sum'}).reset_index()

# создадим объединённую таблицу с заказами и расходами
report_cac = orders_with_sources.merge(costs_grouped_new, left_on=['source_id', 'buy_month'], right_on=['source_id', 'cost_month'])
report_cac = report_cac.groupby(['source_id', 'cost_month']).agg({'costs':'mean', 'uid':'nunique'}).reset_index()

# посчитаем CAC
report_cac['cac'] = report_cac['costs'] / report_cac['uid']

# сформируем итоговый вывод
output_cac = report_cac.pivot_table(index='source_id', columns='cost_month', values='cac')
```

#### Рассчитаем ROMI по когортам в разрезе источников

```python
# подготовим данные для когортного анализа
orders_romi = orders.join(first_sources, on='uid')
orders_romi = orders_romi.merge(costs_grouped_new, left_on=['source_id', 'buy_month'], right_on=['source_id', 'cost_month'])
orders_romi = orders_romi.drop(columns=['buy_ts', 'buy_date'], axis=1)
orders_romi = orders_romi.merge(first_orders, on='uid')

# подготовим данные о первых покупках с данными об источнике трафика
first_orders_with_sources = first_orders.join(first_sources, on='uid')

# посчитаем количество пользователей в каждой когорте с учётом количества трафика
cohort_sizes = (
    first_orders_with_sources.groupby(['source_id', 'first_order_month'])
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['source_id', 'first_order_month', 'n_buyers']

# сформируем когорты
cohorts = (
    orders_romi.groupby(['source_id', 'first_order_month', 'buy_month'])
    .agg({'revenue': 'sum', 'costs':'mean'})
    .reset_index()
)

#сформируем итоговый датасет
report_romi = pd.merge(cohort_sizes, cohorts, on=['source_id', 'first_order_month'])
```

```python
# маржинальность 100%
margin_rate = 1

# рассчитаем валовую выручку и возраст когорт
report_romi['gp'] = report_romi['revenue'] * margin_rate
report_romi['age'] = (
    report_romi['buy_month'] - report_romi['first_order_month']
) / np.timedelta64(1, 'M')
report_romi['age'] = report_romi['age'].round().astype('int')
```

```python
# рассчитаем LTV
report_romi['ltv'] = report_romi['gp'] / report_romi['n_buyers']

# рассчитаем CAC
report_romi['cac'] = report_romi['costs'] / report_romi['n_buyers']

# рассчитаем ROMI
report_romi['romi'] = report_romi['ltv'] / report_romi['cac']

# сформируем итоговый вывод
output_romi = report_romi.pivot_table(
    index=['source_id','first_order_month'], columns='age', values='romi', aggfunc='mean'
).cumsum(axis=1)
```
